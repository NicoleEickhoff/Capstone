# Bayes Time: 

You should compare two populations of your choosing using Bayesian inference. We want to know if the amount of terror one area has seen differs in a significant way than another area (or time period!)

μ average attacks and σ variation across populations 

I will compare suicide hijackers and non-suicide hijackers after 9-11.  My prior will be μ average attacks and σ variation of all hijackers before 9-11. 

We know that suicide hijacking is not recorded before 9-11, so we expect a difference in the two groups-- but maybe not as much as I believe...who knows? After such a "successful" suicide hijacking, terrorists may use fighters as a tool more for this purpose than regular bombings.  Also, 9-11 could have some effect on the non-suicide hijackers - perhaps they use other means to achieve their goals such as kidnapping due to higher security after 9-11.  Presumably, non-suicide hijackers' motive is to obtain something, thus the hightened security may be a deterrent for these terrorists.

In short, we know that hijacking was a non-suicide terror tool pre-9/11, and that it became a suicide-attacker's tool post-9/11 - BUT, did the two groups remain statistically distinct groups (a case in which non-suicide hijakers remain the overwhelming majority) --or did they become more similar (meaning that there are relatively similar levels of each type of hijacking) - or did it even flip (suicide hijackers being the norm while non-suicide hijakers tapered off)?
